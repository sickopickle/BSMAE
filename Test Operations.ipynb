```python
import os
import random
import shutil

def split_data(src_folder, train_folder, eval_folder, test_folder, train_pct, eval_pct):
    # Create destination folders if they don't exist
    for folder in [train_folder, eval_folder, test_folder]:
        os.makedirs(folder, exist_ok=True)
    
    # Get a list of all image files
    all_images = [f for f in os.listdir(src_folder) if os.path.isfile(os.path.join(src_folder, f))]
    num_images = len(all_images)
    
    # Shuffle the list of images (optional)
    random.shuffle(all_images)
    
    # Calculate the number of images for each set
    num_train = int(num_images * train_pct)
    num_eval = int(num_images * eval_pct)
    num_test = num_images - num_train - num_eval
    
    # Move images to their respective folders
    for i, image_file in enumerate(all_images):
        src_path = os.path.join(src_folder, image_file)
        if i < num_train:
            dest_folder = train_folder
        elif i < num_train + num_eval:
            dest_folder = eval_folder
        else:
            dest_folder = test_folder
        dest_path = os.path.join(dest_folder, image_file)
        shutil.move(src_path, dest_path)

# Example usage:
path = "G:/My Drive/CTdata/"
src_folder = path + "AllData"
train_folder = path + "train"
eval_folder = path + "eval"
test_folder = path + "test"
train_pct = 0.7
eval_pct = 0.15

split_data(src_folder, train_folder, eval_folder, test_folder, train_pct, eval_pct)
```
